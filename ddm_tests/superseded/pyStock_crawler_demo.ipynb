# pystock-crawler Demo

There are three functions to demo from pystock-crawler:

1. pystock-crawler symbols grabs ticker symbol lists
1. pystock-crawler prices grabs daily prices
1. pystock-crawler reports grabs fundamentals

Unfortunately, it looks like pystock-crawler isn't compatible with python 3. NOTHING OF VALUE WAS LOST. We can recreate from other sources.

## symbols

pystock-crawler symbols can be easily recreated with the finsymbols package. Please note that this also pulls great sector data.

This could replace the universe file. We'll use a clean definition of sector returns, but we'll gain a larger universe (especially with NYSE and NASDAQ).

```python
import finsymbols

sp500 = pd.DataFrame(finsymbols.get_sp500_symbols())

sp500.head()
```
